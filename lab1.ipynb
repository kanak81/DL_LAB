```python
# 1(a) Tensor Initialization and Data Types
%pip install torch
import torch

# 1D tensor
a = torch.tensor([1, 2, 3], dtype=torch.float32)
print(a)

# Zeros and Ones
zeros = torch.zeros((2, 3))
ones = torch.ones((3, 2))
print(zeros)
print(ones)

# Random tensor
rand = torch.rand((2, 2))
print(rand)

# Empty tensor
x = torch.empty(4, 2)
print(x)

# Linspace tensor
y = torch.linspace(0.1, 5, 15)
print(y)

# Identity Matrix
b = torch.eye(5, 5)
print(b)

# Arange tensor
c = torch.arange(8)
print(c)

# Normal distributed tensor
z = torch.empty(5, 12).normal_(mean=0, std=1)
print(z)

# Different data type conversion
tensor = torch.arange(10)
print(tensor.bool())   # Convert to boolean
print( tensor.short())   # Convert to int16
print(tensor.long())   # Convert to int64
print(tensor.half())   # Convert to float16
print(tensor.float())   # Convert to float32
print(tensor.double())   # Convert to float64
```

```python
# 1(b) Tensor Operations (Arithmetic, Broadcasting, Indexing, Reshaping)

x = torch.tensor([[1, 2], [3, 4]])
y = torch.tensor([[5, 6], [7, 8]])

# Arithmetic operations
print(x + y)
print(x * y)
print(x - y)
print(x / y)

# Broadcasting
z = torch.tensor([10, 20])
print(x + z)

# Indexing
print(x[0])      # First row
print(x[:, 1])   # Second column

# Reshaping
a = torch.arange(6)
a = a.reshape(2, 3)
print(a)
```

```python
# 1(c) Automatic Differentiation using Autograd
a = torch.tensor(2.0, requires_grad=True)
b = a ** 2 + 3 * a + 1

b.backward()
print("Gradient:", a.grad)
```
